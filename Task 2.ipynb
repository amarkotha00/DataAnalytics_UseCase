```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('/Iris.csv')
data.head()
```

```python
X= data.iloc[:, [0, 1, 2, 3]].values

SSE=[]

Krange=np.arange(1,11)

for k in Krange:
  KM=KMeans(n_clusters=k)
  KM.fit(X)
  SSE.append(KM.inertia_)
```

```python
SSE
```

We use the Elboe method to determine the number of clusters needed in this Problem

```python
plt.plot(Krange,SSE)
plt.title('Elbow Curve')
plt.xlabel('Number of clusters _ K')
plt.ylabel('Sum of Squared Error') # Within cluster sum of squares
plt.show()
```

Now We Apply the K means Algorithm by choosing the Number of Clusters to be 3

```python
kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(X)
```

```python

plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=50, c='red',label='Iris-setosa')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=50, c='blue',label='Iris-versicolour')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=50, c='green',label='Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')
```
