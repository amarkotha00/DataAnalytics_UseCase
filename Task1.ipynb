1. **Importing all the modules** 

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

2. **Reading the dataset from a link**

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
```

```python
data.head()
```

```python
data.info()
```

3. **Plot the Data in a graph**

```python
data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

4. **Labels and Outputs**

```python
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values  
```

Import sklearn and Train the dataset

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.15, random_state=0) 
```

Regression Line to be shown on the graph

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
line = regressor.coef_*X+regressor.intercept_

plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

5. **Checking the correctness of trained model**

```python
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

6. **Check with own Input**

```python
Hours = 9.25
result = regressor.predict([[Hours]])
print("No of Hours = 9.25")
print("Predicted Score = {}".format(result[0]))
```

```python
Hours2 = 9.8
result = regressor.predict([[Hours2]])
print("No of Hours = 9.8")
print("Predicted Score = {}".format(result[0]))
```

```python
print("Accuracy is :", regressor.score(X_test, y_test))
```
